## 1. 路亚基本原理

路亚（lure fishing, ルアーフィッシング）是对一类钓鱼技法的总称。这种钓鱼技法起源于美国，*可能是错误的理解* 这种技法最早的目标鱼便是来自北美本土的鱼种，大名鼎鼎的 ~~V (bushi)~~ 大嘴黑鲈， big mouth black bass, 日语：ブラックバス。
不同于大部分未曾接触过钓鱼的人的想象，路亚最大的特点是利用假饵（拟饵），引诱鱼的攻击，而不是通过动植物或者饵料等引诱鱼在捕食的同时咬住钩。

### 1.1 为什么鱼咬不是食物的东西？

不同于最早针对大嘴黑鲈开发的钓法（后简称为鲈钓），在海边大部分常用的鱼饵（lure, ルアー）大多是用铁片制成的，所以每一个刚入门的人都一定会有一个疑问，鱼真的有这么傻么，会去咬铁片？

*大部分情况下，鱼不会咬普通的铁片*

这句话意思是，在水边随便扔个铁片下去，什么也不干的情况下，鱼上钩的概率是微乎其微。所谓路亚，就是通过一切手段，增加鱼咬上铁片的几率。这些手段包括不限于控饵，打窝（提前在目标水域扔下饵料达到聚鱼的效果），船钓，升级装备等等。**但是所有这些手段，归根结底不过是想鱼所想，让鱼主动攻击鱼饵**。为了能够想鱼所想，首先需要思考的便是鱼在什么情况下会咬饵.

**A. 捕食**

最常见同时也是最好理解的原理，就是诱发鱼的捕食。那这时候肯定有人要问了，鱼为什么会捕食铁片啊。这个问题换个方式就是一个究极问题，究竟是吃屎味的巧克力还是巧克力味的屎。

对于一个普通人来说，肯定可以通过嗅觉或者别的痕迹来判断到底是巧克力还是屎。但是假如我们封闭一个人大部分的感官，紧紧留下触觉或者视觉~~（禁止奇怪xp）~~，这个人在吃下肚子之前还能判断吃下的到底是巧克力还是屎么？不排除有一部分天赋异禀的人能通过有限的感官做出争取的判断，但是这时候做出正确判断的概率仍然会大大下降。

相对应的，路亚的目标就是把所谓的**铁片**尽量伪装成**屎味的巧克力**，让鱼心甘情愿的上钩。

**B. 攻击**

这是一个一般情况下很容易被忽视的因素。家里有养宠物的人可能会有类似的经验，动物在表达对某个东西的讨厌和厌烦的情绪的时候，会和人类一样试图攻击这个东西，比如说一个狮群内不可能同时存在两只强壮的狮子，两只狮子一定会互相攻击，再比如说你家里出现一个小偷的时候，除了先报警，~~还可以尝试制服小偷~~。同样的道理也适用于鱼。有的鱼或者有的鱼种处于某些特别时期的时候，会攻击所有出现在视野里有威胁的东西。前一类常指一些领地意识很强的鱼，后一类则常指在排卵期时位于产床附近的鱼类。对于鱼来说，攻击的手段， ~~除了用尾巴给你大逼兜子~~ 一般是用嘴巴对目标进行撕咬，正中我们诱鱼的下怀。

**C. 好奇等**

当然**子非鱼，安知鱼之乐**这句话并没有错误，有时候确实很难理解为什么鱼会对在水中漂浮或者滑行的鱼饵产生兴趣，但是对于路亚人来说鱼对饵有兴趣是一件再好不过的事情。在距离海岸较远的外海进行作钓的时候，偶尔能看见某些大型鱼会追着饵跑却不咬饵，这便是吸引到了鱼的兴趣却没有引起鱼攻击欲望的时候。运气好的时候能在鱼靠近饵极近的timing，猛的拽饵让钩子挂上鱼。

### 1.2 道理我都懂，但是我还是总是钓不上鱼？

这个问题常出现在技术瓶颈期的时候。某些时候你会觉得自己的鱼饵又真实又好看 ~~我是鱼我都会咬一口~~ ，控饵技术也没有什么大的问题，但是为什么鱼就是不上钩？

从我个人丰富的空军经验来看，原因大多有以下几点：

**A. 水域没有鱼**

听起来有点搞笑但是确实是最常见的原因。大部分人有一点误解的是，水边一定会有鱼，但是没有考虑到鱼也是有自己的活动规律的。举例来说，在大海中就有被称为*Oceanic Pole of Inaccessibility* （难抵极）的地区。在太平洋的难抵极附近，由于缺少来自于温暖洋流带来的营养物质，几乎是一片生命禁区。同样的道理也适用于淡水，比如说春季的产卵季，黑鲈就会更集中在产床附近，此时别的水域鱼密度就会下降，鱼会变得更不好钓。

**B. 控饵急躁**

常一起出现的还有嘴硬和摆烂。需要记在心底的是，对于一个普通的钓鱼人，除非是某些时候水里只有鱼没有水的管理场，大部分情况下钓鱼的主基调还是空军。说起来是老生常谈，但是长时间空军的时候往往需要提醒自己不要急躁。可能自己意识不到，但是在急躁的时候往往会同时出现技术动作变形。一定要耐下心来，等待中鱼的窗口期。

**C. 是时候多思考一下了！**

对于入门者来说，往往在对一片水域不够熟悉或者技术手段不够熟练的时候，钓鱼的第一步就是模仿身边的人。但是有一个问题是，这样的模仿一定是对的吗？当然，一般情况下本地人会对所在水域更熟悉，对于什么样的鱼饵什么样的操作更容易中鱼心知肚明，但是这样的操作一定是正确的么？在某些时候思考一下今天的天气或者水文情况下，这样的饵或者操作是否可以做点改变，也许会有惊喜。


## 2. 在第一次路亚前，我需要准备些什么？

终于可以不看前面的那些废话啦！

- *一旦打算把钓鱼作为人生的一个长期兴趣爱好，建议先集中购买一到两套高配装备，尽量减少后期装备升级损失的金钱或者时间*

### 2.1 装备分类

路亚装备的复杂程度可能是劝退大部分人的第一道门槛。但是简单来说，路亚装备可以被归类为以下几类：

- 渔竿
- 渔轮
- 鱼饵
- 线上配件
- 个人配件

这些分类是个人出于经验进行分类的，并不保证一定正确或者泛用。后面的介绍中，对于所有入门第一钓一定要准备的装备，我会用加粗以及类似
***（第一钓）*** 
的字眼进行提醒。

在具体介绍之前，先给出一套具体的适合抛投的入门的装备：

- 渔竿：直柄杆，长度七尺八以内，硬度为M或是L
- 渔轮: 大小为3000左右的纺车轮
- 鱼饵：根据具体作钓场合，e.g., 海水岸抛：21g左右的米诺，30g左右的铁板，10cm左右的波爬；鲈钓：10g左右的米诺
- 线上配件：300m 1.2号pe线, 16lb 前导碳线

抽铁板的情况下对应的入门装备：
- 渔竿：近海铁板杆，即对应铒重在200g以内，同时与渔轮相对应的直柄或者枪柄杆
- 渔轮：根鱼等较小的鱼：线容量为0.8pe 300m的渔轮； 青物等有大型鱼的场合：线容量为3 pe 250m的渔轮
- 鱼饵：根据情况，28g-210g的铁板
- 线上配件：参考渔轮，前导：0.8pe - 12lb前导，3pe - 40lb 前导

同时：
- 个人配件：遮阳装备，路亚钳，路亚用手套

***第一钓前必须准备***

#### 2.2.1 渔杆的参数 之 商品代号
在节2.2的开头中，我们举了一个例子为*AGS 130M-4*，一个类似这样的用字母和数字组成的字符串即为商品代号，往往言简意赅的提供一些厂商想要着重的商品特点；同时正因为如此，导致并不是所有的商品代号都可以一言概之。但是感谢大部分厂商的良心，商品代号往往有以下的信息：
- 适用的渔轮种类 
- 渔杆长度
- 渔杆硬度/灵敏度
- 渔杆节数

### 2.3 渔轮

***第一钓前必备***

在对渔杆的介绍中，我们注意到渔杆有两种对用的渔轮种类，即

- *s, spinning reel, スピニング リール,  纺车轮*
- *b, bait reel, ベイト リール, 水滴轮*

和渔杆比较起来，渔轮的分类要更加简洁明了，只需要保证渔轮和渔杆种类对应即可，即
***纺车轮对应的渔杆必须为直柄杆（s），水滴轮对应的渔杆必须为枪柄杆（b）***

***入门时务必选择互相合适的渔轮和渔杆！！！！***

#### 2.3.1 纺车轮

各个厂商的纺车轮分类异曲同工，大致根据以下因素进行分类

- 淡水或海水 
- 轻型或者重型
- 大小

我们以*daiwa*的旗舰系列*saltiga ソルティガ* 以及 *exist　イグジスト* 为例
[saltiga](https://www.daiwa.com/jp/product/pwzca8e)
![saltiga](pic\5.png)
[exist](https://www.daiwa.com/ja/fishing/item/special/product/exist/exist_lt/lineup.html)
![exist](pic\6.png)

首先可以看见和渔杆类似，商品代号同样在表的最右侧。以 *exist LT2000s-p* 和 *saltiga 8000p* 为例：

- LT: LT是 *daiwa* 独有的一种标号，即light，tough, 保证轻量同时保证刚性
- 2000 或 8000：渔轮的大小。由于构造的问题，这个数字同时直接关系到渔轮的上线长度，可以比较参数 *標準巻糸量 PE号-m* 下的 0.4-200 或者 3-400. 在前半的 0.4 或者 3 是指代pe的号数，后半的 200 或者 400 指对应该号数的可卷上的最大长度。
- s: 一般出现在数字后的s或者d指代 shallow 或者 deep，即纺车轮线杯的深度，s即浅线杯，对应更少的上线量；d即刚好相反。
- p：对应卷线速度 *巻取り長さ*，即手柄每转一圈能回收的线的长度，一般卷线速度 xh > h > 空 > p

在后续参数中，需要格外注意的为：

- 最大ドラグ力（kg）：渔轮可提供的最大拖拽力。需要注意的是，该重量可以在实钓中进行调整，并且由于纺车轮的设计问题，实际上纺车轮可提供的最大拖拽力往往比该数值更小
- 自重（g）：渔轮的重量，~~细狗酌情选择~~
- *標準巻糸量 PE号-m*： ***推荐的线的号数务必与渔杆相匹配***

***入门时选择的纺车应为海水可用的，大小为3000到4000的，卷线速度为H的纺车轮！！！***


#### 2.3.2 水滴轮

简单的来说，除了上述的淡水和海水区分，水滴轮根据能否抛投以及对应的场合，划分为

- 水滴轮
- 铁板轮

具体使用场景上泛用水滴轮一般也可以应对一些60g以下的抽铁板的需要，但是 ***铁板轮一般情况下绝不能用于抛投！！！***

这里需要特别说明的的是，一般情况下泛用常用的纺车轮并没有专门区分抛投或者铁板专用，一般根据自己需要选取适合的大小即可，毕竟如果用一个大小为2000，拖拽力仅为5kg的纺车轮抽铁板钓青物多少都有点不合适吧。。。

##### 2.3.2.1 水滴轮使用

相比于纺车轮，水滴轮更长用于精准抛投，尤其是面对一些较为复杂的抛投场景，比如说树下等抛投空间有限的场合，此时如果用水滴轮配合一跟较软的杆子（腰力为L），我称之为绝杀！但是这个话并不是绝对的，因为结构问题，水滴轮常出现问题 *Back Lash (バックラッシュ)* ，也就是线杯的转速与出线速度不匹配导致炸线。这种不稳定性导致新手上手相对较难并且在外海等极端环境下的不可靠，所以 ***入门请务必选择纺车轮！***

##### 2.3.2.2 水滴轮参数

与纺车轮相似，在选择水滴轮时也需要关注最大ドラグ力（kg）等参数，并且购买与 *標準巻糸量* 对应的枪柄杆。此处以 *SHIMANO Calcutta Conquest (カルカッタコンクエスト) DC* 为例:

[SHIMANO CCQ DC](https://fish.shimano.com/ja-JP/product/reel/baitlurecasting/a075f00002xcal5qac_p.html)
![SHIMANO CCQ DC](pic\7.png)

与纺车轮不同，水滴轮大多使用数字100, 150, 200等表示渔轮的大小，以 *101HG* 为例：

- 100: 101 的 100 指水滴轮大小为100
- 1：101 的 1 指此轮的把手在左侧，即需要用左手操作把手
- **糸巻量ナイロン(lb-m)**： 需要注意与纺车轮不同的是，水滴轮中淡水专用的数量要更多。虽然现在大多数的水滴轮都是海水可，但是这里的**糸巻量ナイロン(lb-m)**还是说明该产品线在最早设计时针对的用线是 *尼龙，Nylon, ナイロン*,也就是说淡水使用的场景。同时与pe线不同的是，尼龙以及碳线大多是通过磅数（lb）而非号数进行区分的。

#### 铁板轮

顾名思义，铁板轮是设计为抽铁板专用的。至于为何铁板轮需要专门的产品线，是因为jig类型的钓法在海钓中几乎占了90%，仅在某些资源极好或者洄游路线上的海域中作钓是才有使用抛投的必要。此外在中深海的jig钓法中，用到的铒重往往在250g以上，对应水深在100m以上，而由于水滴轮的设计，使得该类渔轮的拖拽力在传递到线的过程中损失极少，相比于纺车轮往往在体积更小作钓人更为轻松的情况下就能达到相似的效果。

铁板轮根据本身结构可以分为以下几种：
- 有无线规
- 有无计线器
- 有无通电口

### 2.4 鱼饵，线上配件与个人配件

~~终于写到最后一个部分了~~

正如我们在刚开始提到的，鱼饵决定了能否成功吸引到鱼的注意力，这一点在路亚中格外重要。但是不幸的是，正因为其重要性，鱼饵的种类繁杂如星辰，入门一年半之后的我仍不能说完全掌握了所有鱼饵的使用方法，甚至连认全所有类型的鱼饵也不能有自信说能做到。但是有一点心得如下：

- ***白嫖带你入门的朋友***

刚入门，一定学会白嫖，要大胆的谦逊的同时充满学习精神的白嫖 ~~借鉴绝不是抄袭~~。

- ***多练习***

可以说，每一种鱼饵根据其设计和目标鱼都有对应的控铒技巧，如何让鱼饵的水中姿态最为符合设计初衷，最接近真实的鱼，需要大量且细致的练习，例如抬杆抽铒收线的速率，竿稍的移动方向，铒在水中运动的节奏，等等等等，这些不是想象就可以获得的技巧，不可能一次不练习就能完美的控铒

- ***思考与尝试***

我在开头所说，在始终没有鱼的时候，倘若不思考正在使用的铒使用的钓法是不是符合目前的水文情况，就火急火燎的更换新的铒，也是不合适的。有的时候也要对自己精心挑选的铒充满信心。

至于线上配件，实际上鱼线的选用应当在装备的最开始就决定好。正如节2.2和2.3中反复强调的，渔杆渔轮鱼饵以及线组的选择需要相互配套，具体来说就是选择的线组和鱼饵应当在渔轮和渔杆的参数范围内。

***入门建议同时准备主线和前导线***

原因是主线一般选择pe，前导线一般选择为碳线，在海钓过程中，线组的前端因为需要与海床或鱼接触从而往往极易磨损，而碳线相比pe线要更为耐磨。一般前导线的长度不超过渔杆的长度最佳。同时前导线的强度，即磅数，一般不超过主线。磅数与pe号数的换算，在pe线包装上没有标明强度的情况下，一般以号数乘20或者15进行估算，例如3号pe，估算强度为60lb或者45lb，所以一般选择40lb前导搭配。
