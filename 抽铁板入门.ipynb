*ver 20231224*

**写在开头**

本入门是对入门文档的一点补充，主要针对是jigging，也就是抽铁板这一钓法的内容上的补充。为了便于在日本的同学使用，我会在一些专用词上尽量使用中英日三语对照。说是专用词其实只要看的懂jigging = 抽铁板 = ジギング, 就可以了。出于个人习惯，我常用的是jigging来代指抽铁板。

**本文面对的对象**

- 第一次或是只有过一两次出海jigging的少经验者
- 对渔具有一定了解，自己购买过渔具或者对入门文档中的渔具部分的内容有一定了解的初入门者
- 对于jigging的经验者欢迎对文档查漏补缺
- 主要作钓水域在大阪附近，即西到明石东至三重北抵舞鶴南达串本的地理范围内

**本文不会涵盖的内容**

- 使用真铒的钓法
- 一系列使用串钩和重铅坠， e.g., サビキ和重さ, 的组合钓法
- 拖钓
- 筏钓 （カセ釣り）等出海却不抛投但也不使用铁板的钓法

## 1. 懒人包

对于所有入门jigging的钓鱼爱好者来说，第一步也是最大的问题是渔具挑选。限于我本人的作钓范围目前集中在大阪附近，我能给出的钓组推荐也仅适合大阪附近的水域。对于不涵盖在本文范围内水域的入门者，可以酌情使用比推荐更为重型或轻型的装备。推荐的钓组以及对应的钓法如下：

**快抽**

- 目标鱼代表为青物，即以鰤鱼（ブリ）为代表的掠食型鱼类。
- 铒重：150g-250g
- 渔杆：对应铒重范围在200g附近即可
- 线组：主线：pe1.5 - 3，最大不宜超过3号，下限根据自己需求发挥，建议不低于1.5，长度200m-300m； 前导线：20-40lb，与主线的号数对应
- 渔轮：与线组对应容量的渔轮，一般为5000以上的纺车轮，*shimano*铁板轮1500以上， *daiwa*铁板轮15或300以上

**慢抽**

- 基本所有鱼种都适用
- 铒重：根据不同水域从80g-250g都可使用，个人最常用为120g-180g
- 渔杆：最适铒重在150g附近
- 线组：主线pe0.8 - 2.5，前导线与主线对应即可。
- 渔轮：推荐5000以上的纺车轮或与线容量对应的铁板轮，不宜过于重型

**SLJ**

- 珊瑚区或是底栖鱼较多的水域
- 铒重：根据地区及水文条件的不同，20g-150g都有使用的机会。串本等水流较大地区常用80g以上，在三重等水流更缓的地区常用60g以下
- 渔杆：不推荐较硬的渔杆，对应铒重在100g附近即可
- 线组：pe1.5以下
- 渔轮：纺车轮4000以下，*shimano*铁板轮200以下， *daiwa*铁板轮150以下
